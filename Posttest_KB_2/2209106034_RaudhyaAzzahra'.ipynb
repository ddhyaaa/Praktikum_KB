5. Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

```python
# Konversi dataset ke dalam DataFrame
df = pd.DataFrame(dhya)

# Menghitung informasi yang diperlukan
jumlah_record = len(df)
jumlah_attribute = len(df.columns)
nama_attribute = df.columns.tolist()
jumlah_attribute_angka = len(df.select_dtypes(include=['int64', 'float64']).columns)

# Membuat tabel PrettyTable
table = PrettyTable(['Informasi', 'Nilai'])
table.add_row(['Jumlah Record', jumlah_record])
table.add_row(['Jumlah Attribute', jumlah_attribute])
table.add_row(['Nama Attribute', '\n'.join(nama_attribute)])
table.add_row(['Jumlah Attribute Dengan Tipe Angka', jumlah_attribute_angka])

# Menampilkan tabel
print(table)
```

7. Menampilkan jumlah label dari semua attribute yang bernilai object

```python

# Memilih kolom yang memiliki tipe data objek
kolom_objek = df.select_dtypes(include=['object'])

# Menghitung jumlah label unik dalam setiap kolom objek
jumlah_label = kolom_objek.nunique()

# Menampilkan jumlah label unik
print("Jumlah label unik dalam setiap kolom objek:")
print(jumlah_label)

# Menampilkan nilai-nilai unik dalam setiap kolom objek
print('')
for kolom in jumlah_label.index:
    print(f"Nilai dalamm {kolom}:")
    print(kolom_objek[kolom].unique())
    print('')
```
